```python
import re
import pandas as pd

def normalize_district(name):
    if pd.isna(name):
        return None

    name = name.lower().strip()

    # Remove special characters, brackets, stars
    name = re.sub(r"\(.*?\)", "", name)
    name = re.sub(r"[^a-z\s]", " ", name)

    # Remove extra spaces
    name = re.sub(r"\s+", " ", name).strip()

    return name.title()
```
